```python
#read in csv
df = pd.read_csv('valladolidA.csv')
```

```python
df
```

```python
#filtering dataframe
df = df[df['teamId']=='Barcelona']
```

```python
df
```

```python
#filter the data frame for Barca's passes
df['passer'] = df['playerId']
df['recipient'] = df['playerId'].shift(-1) 
#this above will add the recipient of pass and the original passe
# in the same row of dataframe by creating 2 extra fields

#now we filter the passes from df(successful ones)
passes = df[df['type']=='Pass']
successful = passes[passes['outcome']=='Successful']
```

```python
successful
```

```python
subs = df[df['type']=='SubstitutionOff']
subs = subs['minute']
firstSub = subs.min()
```

```python
firstSub
```

```python
successful = successful[successful['minute']<firstSub]
```

```python
successful
```

```python
pas = pd.to_numeric(successful['passer'],downcast = 'integer')
red = pd.to_numeric(successful['recipient'],downcast = 'integer')
#this just downcasts the float values(jersey numbers) to integers

successful['passer'] = pas
successful['recipient'] = red
```

```python
successful
```

```python
average_locations = successful.groupby('passer').agg({'x':['mean'],'y':['mean','count']})
#group this dataframe together by the same jersey number and give avg val of x and avg val and count of y

average_locations.columns = ['x','y','count']
```

```python
average_locations
```

```python
pass_between = successful.groupby(['passer','recipient']).id.count().reset_index()
pass_between.rename({'id': 'pass_count'}, axis= 'columns', inplace = True)

pass_between = pass_between.merge(average_locations, left_on= 'passer', right_index = True)
#like a Left join on passer b/w two tables
pass_between = pass_between.merge(average_locations, left_on= 'recipient', right_index = True, suffixes = ['','_end'])
#a left join on recipient b/w two tables although col names are same so for rec, suffix has been added
```

```python
pass_between
```

```python
#to add filter based on pass count
pass_between = pass_between[pass_between['pass_count']>3]
```

```python
pass_between
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='#22312b', line_color='#c7d5cc')
fig, ax = pitch.draw(figsize=(16, 11), constrained_layout=False, tight_layout=True)
fig.set_facecolor('#22312b')
#above creates pitch

arrows = pitch.arrows(1.2*pass_between.x, .8*pass_between.y, 1.2*pass_between.x_end, .8*pass_between.y_end, ax = ax,width = 3,headwidth = 3, color = 'yellow', zorder = 1, alpha = .5)
#this is for plotting the arrows

nodes = pitch.scatter(1.2*average_locations.x,.8*average_locations.y,s = 600, color = 'red', edgecolors = 'black' , linewidth = 2.5 , alpha = 1, zorder = 1, ax = ax)
```
